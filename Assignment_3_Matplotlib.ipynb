## ❓ Question 1 — Line Chart: Alphabet Inc. Financial Data

**Task:** Write a Python program to draw line charts of the financial data of **Alphabet Inc.** between **October 3, 2016 to October 7, 2016**.

**Chart Type:** Multi-line chart showing Open, High, Low, Close prices.

````python
# ── Sample Financial Data (fdata.csv) ──────────────────────────────────────
fdata_csv = """Date,Open,High,Low,Close
10-03-16,774.25,776.065002,769.5,772.559998
10-04-16,776.030029,778.710022,772.890015,776.429993
10-05-16,779.309998,782.070007,775.650024,776.469971
10-06-16,779,780.47998,775.539978,776.859985
10-07-16,779.659973,779.659973,770.75,775.080017"""

reader = csv.DictReader(io.StringIO(fdata_csv))
dates, open_p, high_p, low_p, close_p = [], [], [], [], []

for row in reader:
    dates.append(row['Date'])
    open_p.append(float(row['Open']))
    high_p.append(float(row['High']))
    low_p.append(float(row['Low']))
    close_p.append(float(row['Close']))

print("Dates :", dates)
print("Open  :", open_p)
print("High  :", high_p)
print("Low   :", low_p)
print("Close :", close_p)
````

````python
plt.figure(figsize=(10, 6))

plt.plot(dates, open_p,  marker='o', label='Open',  color='blue',   linewidth=2)
plt.plot(dates, high_p,  marker='s', label='High',  color='green',  linewidth=2)
plt.plot(dates, low_p,   marker='^', label='Low',   color='red',    linewidth=2)
plt.plot(dates, close_p, marker='D', label='Close', color='orange', linewidth=2)

plt.title("Alphabet Inc. Financial Data (Oct 3–7, 2016)", fontsize=14, fontweight='bold')
plt.xlabel("Date", fontsize=12)
plt.ylabel("Price (USD)", fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

## ❓ Question 2 — Horizontal Bar Plot: Students vs Marks

**Task:** Write a Python program to show **number of students vs marks obtained** using a **Horizontal Bar Plot**.

```
nos   = [2, 9, 20, 25, 30, 39]
marks = [12, 24, 25, 27, 29, 30]
```

````python
nos   = [2, 9, 20, 25, 30, 39]
marks = [12, 24, 25, 27, 29, 30]

plt.figure(figsize=(8, 5))
bars = plt.barh(marks, nos, color='steelblue', edgecolor='black', height=1.2)

for bar, n in zip(bars, nos):
    plt.text(n + 0.3, bar.get_y() + bar.get_height()/2,
             str(n), va='center', fontsize=10, fontweight='bold')

plt.xlabel("Number of Students", fontsize=12)
plt.ylabel("Marks Obtained", fontsize=12)
plt.title("Number of Students vs Marks Obtained", fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

## ❓ Question 3 — Line Chart: Year-Wise Sales (monthsales.csv)

**Task:** Draw a line chart for year-wise sales of first six months (2018–2021).
- Chart Title: **Year-Wise Sales**
- X-axis: **Months**, Y-axis: **Sales**

````python
# monthsales.csv data — first 6 months for 4 years
monthsales = {
    2018: [2500, 2200, 3000, 2800, 2750, 3500],
    2019: [2600, 1800, 2500, 3000, 3200, 2000],
    2020: [2450, 2000, 3000, 3100, 1950, 3550],
    2021: [3250, 3200, 3800, 2800, 2900, 2300],
}
months = [1, 2, 3, 4, 5, 6]
print("Data loaded:", monthsales)
````

## ❓ Question 6 — Scatter Plot + Bar Chart: Population vs Literacy Rate

**Task:** Using data from `data.gov.in`, plot:
- **Scatter Plot** — Population (X) vs Literacy Rate (Y), grouped by Region, marker = diamond `D`, size = √literacy
- **Bar Chart** — Average Literacy Rate for each Region

````python
# State-wise data: (State, Population in millions, Literacy Rate %, Region)
state_data = [
    ('Uttar Pradesh',    199.8, 67.7, 'North'),
    ('Maharashtra',      112.4, 82.9, 'West'),
    ('Bihar',            104.1, 63.8, 'East'),
    ('West Bengal',       91.3, 77.1, 'East'),
    ('Madhya Pradesh',    72.6, 70.6, 'North'),
    ('Rajasthan',         68.6, 66.1, 'North'),
    ('Tamil Nadu',        72.1, 80.1, 'South'),
    ('Karnataka',         61.1, 75.6, 'South'),
    ('Gujarat',           60.4, 79.3, 'West'),
    ('Andhra Pradesh',    49.4, 67.4, 'South'),
    ('Odisha',            42.0, 73.5, 'East'),
    ('Jharkhand',         33.0, 67.6, 'East'),
    ('Kerala',            33.4, 94.0, 'South'),
    ('Assam',             31.2, 73.2, 'East'),
    ('Punjab',            27.7, 76.7, 'North'),
    ('Haryana',           25.4, 76.6, 'North'),
    ('Delhi',             16.8, 86.3, 'North'),
    ('Himachal Pradesh',   6.9, 83.8, 'North'),
    ('Goa',                1.5, 88.7, 'West'),
    ('Uttarakhand',       10.1, 79.6, 'North'),
]

region_colors = {'North': 'blue', 'South': 'green', 'East': 'red', 'West': 'orange'}
print(f"Total states loaded: {len(state_data)}")
````

````python
# ── Scatter Plot ───────────────────────────────────────────────────────────
plt.figure(figsize=(10, 6))

for region, color in region_colors.items():
    pop   = [d[1] for d in state_data if d[3] == region]
    lit   = [d[2] for d in state_data if d[3] == region]
    sizes = [np.sqrt(l) * 10 for l in lit]   # size = sqrt(literacy)
    plt.scatter(pop, lit,
                color=color, marker='D',
                s=sizes, label=region,
                alpha=0.8, edgecolors='black')

plt.title("Population vs Literacy Rate by Region", fontsize=14, fontweight='bold')
plt.xlabel("Population (Millions)", fontsize=12)
plt.ylabel("Literacy Rate (%)", fontsize=12)
plt.legend(title="Region", fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

````python
# ── Bar Chart: Average Literacy Rate per Region ────────────────────────────
region_literacy = defaultdict(list)
for d in state_data:
    region_literacy[d[3]].append(d[2])

regions      = list(region_colors.keys())
avg_literacy = [np.mean(region_literacy[r]) for r in regions]
colors       = [region_colors[r] for r in regions]

plt.figure(figsize=(8, 5))
bars = plt.bar(regions, avg_literacy, color=colors, edgecolor='black', width=0.5)

for bar, val in zip(bars, avg_literacy):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 0.3,
             f'{val:.1f}%', ha='center', va='bottom', fontweight='bold')

plt.title("Average Literacy Rate by Region", fontsize=14, fontweight='bold')
plt.xlabel("Region", fontsize=12)
plt.ylabel("Average Literacy Rate (%)", fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

## ❓ Question 7 — Box Plot: Student Marks (Marks.csv)

**Task:** Plot the marks data using a **Box Plot** and perform a **comparative analysis** of performance in each subject.

| Name | English | Maths | Hindi | Science | Social Studies |
|------|---------|-------|-------|---------|----------------|
| Rishika Batra | 95 | 95 | 90 | 94 | 95 |
| Waseem Ali | 95 | 76 | 79 | 77 | 89 |
| Kulpreet Singh | 78 | 81 | 75 | 76 | 88 |
| Annie Mathews | 88 | 63 | 67 | 77 | 80 |

````python
marks_data = {
    'English':        [95, 95, 78, 88],
    'Maths':          [95, 76, 81, 63],
    'Hindi':          [90, 79, 75, 67],
    'Science':        [94, 77, 76, 77],
    'Social Studies': [95, 89, 88, 80],
}

subjects = list(marks_data.keys())
data     = [marks_data[s] for s in subjects]
colors   = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FF99CC']

plt.figure(figsize=(10, 6))
bp = plt.boxplot(data, labels=subjects, patch_artist=True, notch=False)

for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.8)

plt.title("Comparative Analysis of Student Performance", fontsize=14, fontweight='bold')
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Marks", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

## ❓ Question 8 — Grouped Bar Chart: Scores by Group and Gender

**Task:** Write a Python program to create a bar plot of **scores by group and gender** using multiple X values for Men and Women.

## ❓ Question 9 — Pie Chart & Subplots (Sales CSV)

### Part (a) — Pie Chart: Total Sales per Product
**Task:** Calculate total sale data for last year for each product and show it using a **Pie Chart** displaying % units sold.

````python
# Sales CSV data (representative structure)
sales_csv = """Month,Bathing soap,Face wash,Toothpaste,Shampoo,Moisturizer
1,9000,1500,4000,3000,2000
2,6500,1200,3800,2800,1800
3,9800,1400,4200,3200,2200
4,8000,1350,4100,3100,1900
5,7600,1250,3900,2900,2100
6,7400,1700,4300,3300,2300
7,7200,1150,3700,2700,1700
8,9900,1800,4500,3500,2500
9,8000,1600,4400,3400,2400
10,11000,1900,4600,3600,2600
11,13000,2100,4800,3800,2800
12,14000,1800,4700,3700,2700"""

products = ['Bathing soap', 'Face wash', 'Toothpaste', 'Shampoo', 'Moisturizer']
totals   = {p: 0 for p in products}
monthly_profit = []
rows = list(csv.DictReader(io.StringIO(sales_csv)))

for row in rows:
    row_total = sum(int(row[p]) for p in products)
    for p in products:
        totals[p] += int(row[p])
    monthly_profit.append(row_total)

print("Yearly totals per product:")
for p, v in totals.items():
    print(f"  {p:15s}: {v}")
````

````python
labels  = list(totals.keys())
sizes   = list(totals.values())
explode = [0.05] * len(labels)

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, explode=explode,
        autopct='%1.1f%%', startangle=140, shadow=True)
plt.title("Total Sales per Product (Units Sold per Year %)",
          fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

### Part (b) — Subplots: Bathing Soap & Face Wash Monthly Sales

**Task:** Read Bathing soap & Face wash data for all months and display using **Subplots**.

````python
months_num   = list(range(1, 13))
bathing_soap = [int(row['Bathing soap']) for row in rows]
face_wash    = [int(row['Face wash'])    for row in rows]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

ax1.plot(months_num, bathing_soap, color='black', marker='o', linewidth=2)
ax1.set_title("Sales data of a Bathingsoap", fontsize=12, fontweight='bold')
ax1.set_ylabel("Sales units in number", fontsize=11)
ax1.grid(True, linestyle='--', alpha=0.5)

ax2.plot(months_num, face_wash, color='red', marker='o', linewidth=2)
ax2.set_title("Sales data of a facewash", fontsize=12, fontweight='bold')
ax2.set_ylabel("Sales units in number", fontsize=11)
ax2.set_xlabel("Month Number", fontsize=11)
ax2.grid(True, linestyle='--', alpha=0.5)

plt.xticks(months_num)
plt.tight_layout()
plt.show()
````

**Task:** Read the total profit of each month and show it using a **Histogram** to visualize the most common profit ranges.
